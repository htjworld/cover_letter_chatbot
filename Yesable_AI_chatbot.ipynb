```python
import pandas as pd

# 파일 경로 지정
file_path = "/Users/htj/Downloads/크롤링_프로젝트/crawling_results/cover_letters.jsonl"

# JSONL 파일을 읽어서 데이터프레임으로 변환 (lines=True 옵션 사용)
df = pd.read_json(file_path, lines=True)

# 데이터프레임 출력
print(df)
```

```python
import pandas as pd
import json

# 기존의 df는 이미 생성된 것으로 가정합니다. 여기서는 'main_text'에 있는 데이터를 사용합니다.

# df['main_text']에서 내용을 반복해서 추출
cover_letter_questions = [
    "1. 삼성전자를 지원한 이유와 입사 후 회사에서 이루고 싶은 꿈을 기술하십시오. (700자 이내)",
    "2. 본인의 성장과정을 간략히 기술하되 현재의 자신에게 가장 큰 영향을 끼친 사건, 인물 등을 기술하시기 바랍니다. (※작품 속 가상 인물도 가능) (1500자 이내)",
    "3. 최근 사회 이슈 중 중요하다고 생각되는 한 가지를 선택하고 이에 관한 자신의 견해를 기술해 주시기 바랍니다. (1000자 이내)",
    "4. 지원한 직무 관련 본인이 갖고 있는 전문지식/경험(심화전공, 프로젝트, 논문, 공모전 등)을 작성하고, 이를 바탕으로 본인이 지원 직무에 적합한 사유를 구체적으로 서술해 주시기 바랍니다. (1000자 이내)"
]

# JSONL 형식으로 변환
formatted_data = []
for _, row in df.iterrows():
    entry = {
        "messages": [
            {"role": "system", "content": "cover_letter"},
            {"role": "user", "content": "\n".join(cover_letter_questions)},
            {"role": "assistant", "content": "\n".join(row['main_text'])}  # main_text를 줄로 이어서 저장
        ]
    }
    formatted_data.append(entry)
```

```python
for i in formatted_data:
    print(i)
```

```python
import json

# formatted_data를 JSONL 파일로 저장하는 코드
with open('cover_letter_finetuning.jsonl', 'w') as f:
    for entry in formatted_data:
        f.write(json.dumps(entry) + "\n")
```

```python
import os
from openai import OpenAI

# 환경 변수에서 API_KEY 불러오기
api_key = os.getenv("OPENAI_API_KEY")

# OpenAI 클라이언트 인스턴스 생성
client = OpenAI(api_key=api_key)

# JSONL 파일 업로드
client.files.create(
  file=open("cover_letter_finetuning.jsonl", "rb"),
  purpose="fine-tune"
)
```

```python
job_create = client.fine_tuning.jobs.create(
  training_file="file-u15EdGjrVsiaDNrRv9WboID8",
  model="gpt-4o-mini-2024-07-18"
)
job_create
```

```python
import pandas as pd

# Chatbot friendly and inquisitive responses
df_chatbot = pd.DataFrame({
    'questions': [
        '오늘 날씨는 어때요?', 
        '오늘 저녁 뭐 먹을까?', 
        '주말 계획 있나요?', 
        '어제 드라마 봤어?', 
        '영화 추천해줄래?', 
        '요즘 뭐 재밌는 일 있어?', 
        '내일 뭐 할까?', 
        '새로운 취미 추천해줘', 
        '오늘 하루 어땠어?', 
        '내가 뭘 해야 행복할까?'
    ],
    'answers': [
        '오늘은 맑고 화창한 날씨입니다. 오늘은 어떤 거 할 예정인가요? 😊',
        '오늘 저녁은 따뜻한 국물 요리는 어때요? 아니면 파스타도 좋을 것 같아요! 🍝',
        '주말에는 잠깐 산책하거나 영화 보는 것도 좋지 않을까요? 계획 있으세요? 🎥',
        '어제 드라마 봤어요! 정말 재미있었죠? 그중에 뭐가 제일 인상 깊었나요? 😄',
        '액션 영화 좋아하시면, 최근에 나온 영화 한 번 보세요! 어떤 장르 좋아하세요? 🎬',
        '요즘은 책 읽는 게 재밌더라고요. 새로운 취미 생각해 본 적 있나요? 📚',
        '내일은 가까운 카페에 가거나 친구와 만나보는 건 어때요? 좋은 계획 있나요? ☕',
        '새로운 취미로 요가나 베이킹 추천할게요! 관심 있는 거 있으세요? 🍰',
        '오늘 하루도 바쁘셨죠? 어떤 일이 가장 기억에 남으세요? 😌',
        '행복은 일상의 작은 순간에서 오지 않을까요? 지금 가장 하고 싶은 게 뭐예요? 🌟'
    ]
})

df_chatbot.columns = ['prompt', 'completion']

# Converting to the specified JSON format
formatted_chatbot_data = [
    {
        "messages": [
            {"role": "system", "content": "chat_bot"},
            {"role": "user", "content": row['prompt']},
            {"role": "assistant", "content": row['completion']}
        ]
    }
    for _, row in df_chatbot.iterrows()
]

# Print each formatted JSON entry
for entry in formatted_chatbot_data:
    print(entry)
```

```python
import json

# formatted_data를 JSONL 파일로 저장하는 코드
with open('chat_bot_finetuning.jsonl', 'w') as f:
    for entry in formatted_chatbot_data:
        f.write(json.dumps(entry) + "\n")
```
